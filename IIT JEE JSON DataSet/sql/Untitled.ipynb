```python
from datasets import load_dataset
import json

# Load the dataset
ds = load_dataset("aladrira/sql-mysql-db")

# Convert the 'train' split into a list of dictionaries
data_list = ds['train'].to_list()

# Save everything into a JSON file
with open("sql_mysql_db.json", "w", encoding="utf-8") as f:
    json.dump(data_list, f, indent=2, ensure_ascii=False)

print("✅ Entire dataset saved as sql_mysql_db.json")
```
