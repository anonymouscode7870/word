```python
from datasets import load_dataset
import pandas as pd
import json

# Load the dataset (requires Hugging Face login if private)
ds = load_dataset("CK0607/2025-Jee-Mains-Question")

# Inspect available splits
print(ds)

# Convert to DataFrame (example: 'train' split)
jee_df = pd.DataFrame(ds['train'])

# Save all data into a single JSON file
all_data = jee_df.to_dict(orient="records")

with open("jee_mains_2025_all.json", "w", encoding="utf-8") as f:
    json.dump(all_data, f, indent=2, ensure_ascii=False)

print("✅ JEE Mains 2025 PYQs saved to jee_mains_2025_all.json")
```
