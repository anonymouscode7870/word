```python
from datasets import load_dataset
import json

# Load the dataset (requires HuggingFace login if private)
ds = load_dataset("ruh-ai/grafite-jee-mains-qna-no-img")

# Convert the 'train' split into a list of dictionaries
data_list = ds['train'].to_list()

# Save everything into a JSON file
with open("grafite_jee_mains_qna.json", "w", encoding="utf-8") as f:
    json.dump(data_list, f, indent=2, ensure_ascii=False)

print("✅ Entire dataset saved as grafite_jee_mains_qna.json")
```

```python
import json

# Load the JSON file
with open("grafite_jee_mains_qna.json", "r", encoding="utf-8") as f:
    data = json.load(f)

# Collect unique paper_id values
unique_ids = sorted({item.get("paper_id") for item in data if item.get("paper_id")})

# Print them
for pid in unique_ids:
    print(pid)
```
