```python
from datasets import load_dataset
import pandas as pd
import json

# Load the evaluation split
dataset = load_dataset("Reja1/jee-neet-benchmark", split="test")

# Inspect dataset structure
print(dataset)

# Convert to pandas DataFrame for easier handling
df = pd.DataFrame(dataset)

# Example: Access the first question
example = dataset[0]
image = example["image"]          # PIL Image object
question_id = example["question_id"]
subject = example["subject"]
correct_answers = json.loads(example["correct_answer"])  # Parse JSON string

print(f"Question ID: {question_id}")
print(f"Subject: {subject}")
print(f"Correct Answer(s): {correct_answers}")

# If you want to save the entire dataset into a single JSON file:
all_data = df.to_dict(orient="records")
with open("jee_neet_benchmark.json", "w", encoding="utf-8") as f:
    json.dump(all_data, f, indent=2, ensure_ascii=False)

print("✅ Dataset saved to jee_neet_benchmark.json")
```

```python
from datasets import load_dataset
import os, json

# Load dataset
dataset = load_dataset("Reja1/jee-neet-benchmark", split="test")

# Create folder for images
os.makedirs("images", exist_ok=True)

records = []
for example in dataset:
    # Save image to file (using question_id for unique naming)
    img_filename = f"question_{example['question_id']}.png"
    img_path = os.path.join("images", img_filename)
    example["image"].save(img_path)

    # Build record without raw image object, but with path reference
    record = {k: v for k, v in example.items() if k != "image"}
    record["image_link"] = img_path  # add link to saved image
    records.append(record)

# Save JSON with image links
with open("jee_neet_benchmark.json", "w", encoding="utf-8") as f:
    json.dump(records, f, indent=2, ensure_ascii=False)

print("✅ JSON saved with image links")
```
