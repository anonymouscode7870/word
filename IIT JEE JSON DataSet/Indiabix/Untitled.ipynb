```python
import json
import re

def remove_chinese(text):
    if not isinstance(text, str):
        return text
    # Matches Chinese characters and removes them
    return re.sub(r'[\u4e00-\u9fff]+', '', text)

def clean_json_recursive(data):
    if isinstance(data, dict):
        # Cleans both keys and values
        return {remove_chinese(k): clean_json_recursive(v) for k, v in data.items()}
    elif isinstance(data, list):
        return [clean_json_recursive(item) for item in data]
    else:
        return remove_chinese(data)

# 1. Load your file
with open('indiabix2024.json', 'r', encoding='utf-8') as f:
    original_data = json.load(f)

# 2. Clean the data
cleaned_data = clean_json_recursive(original_data)

# 3. Save as a new JSON file
with open('cleaned_indiabix2024.json', 'w', encoding='utf-8') as f:
    json.dump(cleaned_data, f, indent=4, ensure_ascii=False)

print("Cleaning complete. Saved to 'cleaned_indiabix2024.json'.")
```

```python
import json
import re

# 1. Mapping for parameters you want to keep and rename
KEY_MAP = {
    "问题选项": "options",
    "解析如下": "explanation",
    "答案是": "answer_key",
    "元数据": "metadata"
}

# 2. Fields you want to delete completely
REMOVE_FIELDS = {"create_time", "问答来源", "时间元数据"}

def clean_text(text):
    if not isinstance(text, str):
        return text
    # Removes Chinese characters (\u4e00-\u9fff)
    cleaned = re.sub(r'[\u4e00-\u9fff]+', '', text)
    # Clean up leftover colon prefixes and whitespace
    return cleaned.strip().lstrip(':').strip()

def process_data(data):
    if isinstance(data, dict):
        new_dict = {}
        for k, v in data.items():
            # Skip fields targeted for removal
            if k in REMOVE_FIELDS:
                continue
            
            # Rename if in map, otherwise clean the Chinese out of the key
            new_key = KEY_MAP.get(k, clean_text(k))
            
            # If the key becomes empty after cleaning, give it a placeholder
            if not new_key:
                new_key = "data_field"
            
            new_dict[new_key] = process_data(v)
        return new_dict
    
    elif isinstance(data, list):
        return [process_data(item) for item in data]
    
    else:
        return clean_text(data)

# Execution
try:
    with open('indiabix2024.json', 'r', encoding='utf-8') as f:
        original_json = json.load(f)

    final_output = process_data(original_json)

    with open('cleaned_indiabix2024.json', 'w', encoding='utf-8') as f:
        json.dump(final_output, f, indent=4, ensure_ascii=False)
        
    print("Filtered and renamed JSON successfully.")
except Exception as e:
    print(f"Error: {e}")
```

```python
import json
import re

def clean_content(text):
    if not isinstance(text, str):
        return text
    # Remove Chinese characters
    cleaned = re.sub(r'[\u4e00-\u9fff]+', '', text)
    # Clean up artifacts like leading colons or " :"
    return cleaned.strip().lstrip(':').strip()

def process_node(data):
    if isinstance(data, dict):
        new_dict = {}
        for k, v in data.items():
            # 1. STRICT REMOVAL of only these two
            if k in ["create_time", "问答来源"] or "时间元数据" in k:
                continue
            
            # 2. INTELLIGENT RENAMING
            # We look for keywords in the Chinese keys to rename them properly
            lowered_k = k.lower()
            if "问题" in k or "question" in lowered_k:
                new_key = "question"
            elif "选项" in k or "options" in lowered_k:
                new_key = "options"
            elif "答案" in k or "answer" in lowered_k:
                new_key = "correct_answer"
            elif "解析" in k or "explanation" in lowered_k:
                new_key = "explanation"
            elif "id" in lowered_k:
                new_key = "id"
            else:
                new_key = clean_content(k)
            
            # 3. Handle the value (keep lists/dicts, clean strings)
            if not new_key: new_key = "field"
            new_dict[new_key] = process_node(v)
        return new_dict

    elif isinstance(data, list):
        return [process_node(item) for item in data]
    
    else:
        # If it's the actual text (like the question body), remove Chinese
        return clean_content(data)

# Run the process
try:
    with open('indiabix2024.json', 'r', encoding='utf-8') as f:
        data = json.load(f)

    cleaned_data = process_node(data)

    with open('cleaned_indiabix2024.json', 'w', encoding='utf-8') as f:
        json.dump(cleaned_data, f, indent=4, ensure_ascii=False)
    
    print("Done! Questions, Options, and Answers are preserved and renamed.")
except Exception as e:
    print(f"Error: {e}")
```

```python
import json

# Load the JSON file
with open("indiabix2024.json", "r", encoding="utf-8") as f:
    data = json.load(f)

# Function to rename keys
def rename_keys(record):
    mapping = {
        "问": "question",
        "答": "answer",
        "来源": "source_site",
        "时间": "date",
        "元数据": "metadata"
    }
    new_record = {}
    for k, v in record.items():
        # Rename if in mapping
        if k in mapping:
            new_record[mapping[k]] = v
        else:
            new_record[k] = v

    # Inside metadata, remove unwanted keys
    if "metadata" in new_record and isinstance(new_record["metadata"], dict):
        new_record["metadata"].pop("题目链接", None)
        new_record["metadata"].pop("评论链接(可能含用户解析)", None)

    return new_record

# Apply renaming/removal to all records
clean_data = [rename_keys(item) for item in data]

# Save cleaned JSON
with open("indiabix2024_clean00.json", "w", encoding="utf-8") as f:
    json.dump(clean_data, f, indent=2, ensure_ascii=False)

print("✅ Cleaned dataset saved to indiabix2024_clean.json")
```

```python
import json
import re

# Load the JSON file
with open("indiabix2024.json", "r", encoding="utf-8") as f:
    data = json.load(f)

cleaned_data = []

for item in data:
    new_item = {}

    # Rename Chinese fields
    if "问" in item:
        new_item["question"] = item["问"]
    if "答" in item:
        new_item["answer"] = item["答"]
        # Try to extract correct option from answer text (e.g. "答案是:D")
        match = re.search(r"答案是[:：]\s*([A-D])", item["答"])
        if match:
            new_item["correct_option"] = match.group(1)

    if "来源" in item:
        new_item["source_site"] = item["来源"]
    if "时间" in item:
        new_item["date"] = item["时间"]

    # Handle 元数据 (metadata) if present
    metadata = item.get("元数据", {})
    if isinstance(metadata, dict):
        # Extract options
        if "问题选项" in metadata:
            new_item["options"] = metadata["问题选项"]

        # Extract explanation
        if "解析" in metadata:
            new_item["explanation"] = metadata["解析"]

        # Extract category
        if "问题类别" in metadata:
            new_item["category"] = metadata["问题类别"]

    # Remove metadata entirely (not kept in final JSON)
    # So we don't copy metadata itself

    cleaned_data.append(new_item)

# Save cleaned JSON
with open("indiabix2024_final432.json", "w", encoding="utf-8") as f:
    json.dump(cleaned_data, f, indent=2, ensure_ascii=False)

print("✅ Cleaned dataset saved to indiabix2024_final.json")
```

```python
import json
import re

# Load the JSON file
with open("indiabix2024.json", "r", encoding="utf-8") as f:
    data = json.load(f)

cleaned_data = []

for item in data:
    new_item = {}

    # Question
    if "问" in item:
        new_item["question"] = item["问"]

    # Answer + Correct Option
    if "答" in item:
        raw_answer = item["答"]

        # Extract correct option (e.g. "答案是:B")
        match = re.search(r"答案是[:：]\s*([A-D])", raw_answer)
        if match:
            new_item["correct_option"] = match.group(1)

        # Remove the Chinese prefix "答案是:X 解析如下"
        cleaned_answer = re.sub(r"答案是[:：]\s*[A-D]\s*解析如下[:：]?", "", raw_answer)
        new_item["answer"] = cleaned_answer.strip()

    # Source and Date
    if "来源" in item:
        new_item["source_site"] = item["来源"]
    if "时间" in item:
        new_item["date"] = item["时间"]

    # Metadata cleanup
    metadata = item.get("元数据", {})
    if isinstance(metadata, dict):
        if "问题选项" in metadata:
            new_item["options"] = metadata["问题选项"]
        if "解析" in metadata:
            new_item["explanation"] = metadata["解析"]
        if "问题类别" in metadata:
            new_item["category"] = metadata["问题类别"]

    # Do not keep 元数据 itself
    cleaned_data.append(new_item)

# Save cleaned JSON
with open("indiabix2024_final543.json", "w", encoding="utf-8") as f:
    json.dump(cleaned_data, f, indent=2, ensure_ascii=False)

print("✅ Cleaned dataset saved to indiabix2024_final.json")
```

```python
import json

# Load the cleaned JSON file
with open("indiabix2024_final543.json", "r", encoding="utf-8") as f:
    data = json.load(f)

# Define categories to remove
remove_categories = {"Interview", "Placement Papers", "Accenture"}

# Filter out unwanted categories
filtered_data = []
for item in data:
    if "category" in item and item["category"]:
        # Handle list or string
        if isinstance(item["category"], list):
            if any(cat in remove_categories for cat in item["category"]):
                continue  # skip this record
        else:
            if item["category"] in remove_categories:
                continue  # skip this record
    filtered_data.append(item)

# Save the filtered dataset
with open("indiabix2024_no_interview.json", "w", encoding="utf-8") as f:
    json.dump(filtered_data, f, indent=2, ensure_ascii=False)

print(f"✅ Cleaned dataset saved to indiabix2024_no_interview.json with Interview, Placement Papers, and Accenture removed")
```
