```python
from datasets import load_dataset
import pandas as pd
import json

# Load the dataset
ds = load_dataset("tubeiFu/indiabix2024")

# Convert to pandas DataFrame (example: 'train' split)
df = pd.DataFrame(ds['train'])

# Convert entire dataset into list of dictionaries
all_data = df.to_dict(orient="records")

# Save everything into one JSON file
with open("indiabix2024.json", "w", encoding="utf-8") as f:
    json.dump(all_data, f, indent=2, ensure_ascii=False)

print("✅ Full dataset saved to indiabix2024.json")

```

```python
from datasets import load_dataset
import pandas as pd
import json

# Load dataset
ds = load_dataset("tubeiFu/indiabix2024")

# Convert to pandas DataFrame
df = pd.DataFrame(ds['train'])

# Rename columns from Chinese to English
df = df.rename(columns={
    "问": "question",
    "答": "answer",
    "来源": "source",
    "时间": "time",
    "元数据": "metadata"
})

# Convert to list of dicts
all_data = df.to_dict(orient="records")

# Save to JSON with UTF-8 encoding
with open("indiabix2024.json", "w", encoding="utf-8") as f:
    json.dump(all_data, f, indent=2, ensure_ascii=False)

print("✅ Dataset saved to indiabix2024.json with English field names")
```
