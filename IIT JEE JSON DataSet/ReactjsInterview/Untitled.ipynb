```python
import json
from datasets import load_dataset

# Load dataset
ds = load_dataset("anuragkmr45/reactjs-interview-question")

# Convert to Python list of dicts
data = ds["train"].to_list()

# Save as a valid JSON array
with open("reactjs_interview_questions.json", "w", encoding="utf-8") as f:
    json.dump(data, f, ensure_ascii=False, indent=2)
```
