```python
import kagglehub
import json
import ast
import os

# Download latest version
path = kagglehub.dataset_download("mrutyunjaybiswal/iitjee-neet-aims-students-questions-data")
print("Path to dataset files:", path)

# Example: iterate through all JSON files in the dataset folder
for filename in os.listdir(path):
    if filename.endswith(".json"):
        file_path = os.path.join(path, filename)
        with open(file_path, "r", encoding="utf-8") as f:
            data = json.load(f)

        # --- Clean options ---
        def clean_options(options):
            new_options = {}
            for opt in options:
                try:
                    parsed = ast.literal_eval(opt)
                    if isinstance(parsed, dict):
                        identifier = parsed.get("identifier")
                        content = parsed.get("content", "").strip()
                        if identifier:
                            new_options[identifier] = content
                except Exception:
                    pass
            return new_options

        # --- Clean correct_option ---
        def clean_correct_option(value):
            if isinstance(value, str):
                try:
                    parsed = json.loads(value)
                    if isinstance(parsed, list):
                        return parsed
                except Exception:
                    return [value]
            return value

        def update_all(obj):
            if isinstance(obj, dict):
                for k, v in obj.items():
                    if k == "options" and isinstance(v, list):
                        obj[k] = clean_options(v)
                    elif k == "correct_option":
                        obj[k] = clean_correct_option(v)
                    else:
                        update_all(v)
            elif isinstance(obj, list):
                for item in obj:
                    update_all(item)

        update_all(data)

        # Save cleaned file alongside original
        clean_path = os.path.join(path, f"clean_{filename}")
        with open(clean_path, "w", encoding="utf-8") as f:
            json.dump(data, f, indent=2, ensure_ascii=False)

        print(f"Cleaned file saved: {clean_path}")
```

```python
import csv
import json

# Input and output file paths
csv_file = "subjects-questions.csv"
json_file = "subjects-questions.json"

# Read CSV and convert to list of dicts
with open(csv_file, "r", encoding="utf-8") as f:
    reader = csv.DictReader(f)
    rows = list(reader)

# Save as JSON
with open(json_file, "w", encoding="utf-8") as f:
    json.dump(rows, f, indent=2, ensure_ascii=False)

print(f"Converted {csv_file} → {json_file}")
```
