```python
from datasets import load_dataset
import json

# Load the dataset
ds = load_dataset("AdithyaSNair/cbse-papers-2009-2025")

# Convert the 'train' split into a list of dicts
data_list = ds['train'].to_list()

# Filter only years 2020–2025
filtered_data = [item for item in data_list if 2020 <= int(item.get("year", 0)) <= 2025]

# Save to JSON
with open("CBSE_PAPERS_2020_2025.json", "w", encoding="utf-8") as f:
    json.dump(filtered_data, f, indent=2, ensure_ascii=False)

print("✅ Saved CBSE papers from 2020–2025 into CBSE_PAPERS_2020_2025.json")
```
