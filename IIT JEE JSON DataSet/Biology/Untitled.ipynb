```python
from datasets import load_dataset
import pandas as pd
import json

# Load the dataset
ds = load_dataset("sweatSmile/neet-biology-qa")

# Convert to pandas DataFrame (example: 'train' split)
bio_df = pd.DataFrame(ds['train'])

# Convert entire dataset into list of dictionaries
all_data = bio_df.to_dict(orient="records")

# Save everything into one JSON file
with open("neet_biology_qa.json", "w", encoding="utf-8") as f:
    json.dump(all_data, f, indent=2, ensure_ascii=False)

print("✅ NEET Biology QA dataset saved to neet_biology_qa.json")
```
