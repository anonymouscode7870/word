```python
import pandas as pd
from datasets import load_dataset
import json

# Load the Chemistry dataset
ds = load_dataset("Abc8264/Jee-Chemistry-dataset-with-COT")

# Convert to pandas DataFrame
chem_df = pd.DataFrame(ds['train'])

# Convert the entire dataset into a list of dictionaries
all_data = chem_df.to_dict(orient="records")

# Save everything into one JSON file
with open("chemistry_all.json", "w", encoding="utf-8") as f:
    json.dump(all_data, f, indent=2, ensure_ascii=False)

print("✅ Single JSON file created: chemistry_all.json")
```
