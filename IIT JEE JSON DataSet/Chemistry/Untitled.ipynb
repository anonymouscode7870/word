```python
import pandas as pd
import re
import json
from bs4 import BeautifulSoup
from datasets import load_dataset

# 1. Cleaning Function
def clean_chemistry_text(text):
    if not isinstance(text, str) or text == "null":
        return text
    
    # Handle HTML subscripts and superscripts specifically for Chemistry
    # Converts <sub>3</sub> to 3 or _3 depending on your preference
    text = re.sub(r'<sub>(.*?)</sub>', r'_\1', text)
    text = re.sub(r'<sup>(.*?)</sup>', r'^\1', text)
    
    # Remove all other HTML tags
    soup = BeautifulSoup(text, "html.parser")
    clean_text = soup.get_text()
    
    # Clean up whitespace and LaTeX artifacts
    clean_text = clean_text.replace("~", " ")
    clean_text = re.sub(r'\s+', ' ', clean_text).strip()
    
    return clean_text

# 2. Load and Transform
ds = load_dataset("Abc8264/Jee-Chemistry-dataset-with-COT")
chem_df = pd.DataFrame(ds['train'])

# 3. Apply Cleaning
# Clean the main text fields
chem_df['question'] = chem_df['question'].apply(clean_chemistry_text)
chem_df['solution'] = chem_df['solution'].apply(clean_chemistry_text)

# Fix the 'options' column (it's often a string in these datasets)
def parse_options(opt):
    if isinstance(opt, str):
        try:
            # Replace single quotes with double quotes if necessary for JSON
            return json.loads(opt.replace("'", '"'))
        except:
            return opt
    return opt

chem_df['options'] = chem_df['options'].apply(parse_options)

# 4. Save to JSON
all_data = chem_df.to_dict(orient="records")
with open("chemistry_cleaned.json", "w", encoding="utf-8") as f:
    json.dump(all_data, f, indent=2, ensure_ascii=False)

print("✅ Cleaned JSON created.")
```

```python
import pandas as pd
import re
import json
from bs4 import BeautifulSoup
from datasets import load_dataset

def clean_chemistry_text(text):
    if not isinstance(text, str) or text == "null":
        return text
    
    # 1. Convert HTML tags to a readable format before stripping
    text = text.replace('<sub>', '_').replace('</sub>', '')
    text = text.replace('<sup>', '^').replace('</sup>', '')

    # 2. Extract LaTeX blocks to "protect" them from cleaning
    # This finds everything between $$ and $$
    latex_blocks = re.findall(r'\$\$.*?\$\$', text)
    placeholder = "LATEX_BLOCK_PHOLDER"
    
    # Replace LaTeX with placeholders
    temp_text = re.sub(r'\$\$.*?\$\$', placeholder, text)
    
    # 3. Clean the remaining prose
    soup = BeautifulSoup(temp_text, "html.parser")
    clean_text = soup.get_text()
    
    # Remove excessive tildes and weird spacing
    clean_text = clean_text.replace("~", " ")
    clean_text = re.sub(r'\s+', ' ', clean_text).strip()
    
    # 4. Put the LaTeX blocks back exactly as they were
    for block in latex_blocks:
        clean_text = clean_text.replace(placeholder, block, 1)
    
    return clean_text

# Load and Transform
ds = load_dataset("Abc8264/Jee-Chemistry-dataset-with-COT")
chem_df = pd.DataFrame(ds['train'])

# Apply Cleaning
chem_df['question'] = chem_df['question'].apply(clean_chemistry_text)
chem_df['solution'] = chem_df['solution'].apply(clean_chemistry_text)

# Save to JSON
all_data = chem_df.to_dict(orient="records")
with open("chemistry_cleaned2.json", "w", encoding="utf-8") as f:
    json.dump(all_data, f, indent=2, ensure_ascii=False)

print("✅ Cleaned JSON created with LaTeX preserved.")
```

```python
import json
import ast

with open("chemistry_all.json", "r", encoding="utf-8") as f:
    data = json.load(f)

def clean_options(options):
    new_options = {}
    for opt in options:
        try:
            parsed = ast.literal_eval(opt)  # turn string into dict
            if isinstance(parsed, dict):
                identifier = parsed.get("identifier")
                content = parsed.get("content", "").strip()
                if identifier:
                    new_options[identifier] = content
        except Exception:
            pass
    return new_options

def clean_correct_option(value):
    # If it's a string like '["D"]', parse it into a real list
    if isinstance(value, str):
        try:
            parsed = json.loads(value)
            if isinstance(parsed, list):
                return parsed
        except Exception:
            return [value]
    return value

def update_all(obj):
    if isinstance(obj, dict):
        for k, v in obj.items():
            if k == "options" and isinstance(v, list):
                obj[k] = clean_options(v)
            elif k == "correct_option":
                obj[k] = clean_correct_option(v)
            else:
                update_all(v)
    elif isinstance(obj, list):
        for item in obj:
            update_all(item)

update_all(data)

with open("jee_chemistry_perfected_clean5.json", "w", encoding="utf-8") as f:
    json.dump(data, f, indent=2, ensure_ascii=False)
```

```python
import json

def convert_options_field(entry):
    try:
        # Parse the stringified options array
        options_list = json.loads(entry["options"])
        # Convert into dictionary { "A": "...", "B": "...", ... }
        options_dict = {opt["identifier"]: opt["content"] for opt in options_list}
        entry["options"] = options_dict
    except Exception:
        # If already in dict form or parsing fails, leave as is
        pass
    return entry

def process_file(input_file, output_file):
    with open(input_file, "r") as f:
        data = json.load(f)  # assumes chemistry_all.json is a JSON array

    # Transform each entry
    for i, entry in enumerate(data):
        data[i] = convert_options_field(entry)

    # Save back to new file
    with open(output_file, "w") as f:
        json.dump(data, f, indent=2)

# Example usage:
# process_file("chemistry_all.json", "chemistry_all_fixed.json")
```

```python
import json

# Load your JSON file
with open("chemistry_all.json", "r", encoding="utf-8") as f:
    data = json.load(f)

# Iterate through each record and fix options/correct_option
for item in data:
    # Convert options if it's a string
    if isinstance(item.get("options"), str):
        try:
            item["options"] = json.loads(item["options"])
        except json.JSONDecodeError:
            print(f"Skipping malformed options in question_id {item.get('question_id')}")

    # Convert correct_option if it's a string
    if isinstance(item.get("correct_option"), str):
        try:
            item["correct_option"] = json.loads(item["correct_option"])
        except json.JSONDecodeError:
            print(f"Skipping malformed correct_option in question_id {item.get('question_id')}")

# Save back to file
with open("chemistry_all_fixed.json", "w", encoding="utf-8") as f:
    json.dump(data, f, indent=2, ensure_ascii=False)

print("✅ Finished converting options and correct_option to arrays.")
```
