```python
import pandas as pd
from datasets import load_dataset

# Load January session evaluation set
jan_data = load_dataset("PhysicsWallahAI/JEE-Main-2025-Math", "jan", split="test")

# Load April session evaluation set
apr_data = load_dataset("PhysicsWallahAI/JEE-Main-2025-Math", "apr", split="test")

# Convert Hugging Face Dataset to pandas DataFrame
jan_df = pd.DataFrame(jan_data)
apr_df = pd.DataFrame(apr_data)

# Ensure all required fields exist, even if empty
required_fields = [
    "question",
    "answer",
    "question_type",
    "options",
    "correct_options",
    "additional_data",
    "metadata"
]

for df in [jan_df, apr_df]:
    for field in required_fields:
        if field not in df.columns:
            df[field] = None  # add missing columns with None

# Save as proper JSON arrays
jan_df.to_json("jan_data.json", orient="records", indent=2)
apr_df.to_json("apr_data.json", orient="records", indent=2)
```
