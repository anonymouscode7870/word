```python
import json

def load_data(filename):
    with open(filename, "r") as f:
        data = json.load(f)
    return data

```

```python
data = load_data("store_data.json")
print(data)
print(type(data))  
```

```python
# data cleaning
def view_data(data):
    for user in data:
        print(user)
```

```python
view_data(data)
```

```python
# clean data
def clean_data(data):
    text_to_num = {"one":1, "two":2, "three":3, "four":4, "five":5}

    for users in data:
        raw_rating = users["rating"].strip().lower()
        print(raw_rating)
        if(raw_rating in text_to_num):
            raw_rating = text_to_num[raw_rating]
        users["rating"] = raw_rating

```

```python
clean_data(data)
```

```python
view_data(data)
```
