## Histogram

```python
import matplotlib.pyplot as plt
import numpy as np

# histogram used for numerical data not for the historical data
# random no in array
# This means every time you run your code, you will get the exact same sequence of numbers.
np.random.seed(0)
      # np.random.normal(mean, STD_Devia, Size)
score = np.random.normal(70, 30, 100).astype(int)
score2 = np.random.normal(60, 34, 100).astype(int)
print(score)
print(score2)
```

```python
plt.hist(score, alpha=0.3, edgecolor = "yellow", color="yellow", label="students" )
plt.hist(score2, alpha=0.3, edgecolor = "red", color="blue", label="teachers")
plt.legend()
```

```python
plt.ylabel("No of students")
plt.xlabel("Scores")
plt.hist(score, color="green", edgecolor="red")
```

```python
bins = [0, 20, 40, 60, 80, 100, 120, 140]
plt.ylabel("No of students")
plt.xlabel("Scores")
plt.hist(score, color="green", edgecolor="red", bins=bins)
plt.title("Students Ovrview")
plt.grid("True")

# axvline
plt.axvline(  #vertical line
    x=13,
    color="red"
)

plt.axhline(  # horizontal line
    y=13, 
    color="orange", 
    linestyle="--", 
    label="Threshold"
)
```
