```python
# Movies Data

# Oscar Winner Movies
oscar_movies = [
    "The Dark Knight", 
    "The Hurt Locker",  
    "The King's Speech", 
    "The Artist",     
    "Argo"
]
oscar_years = [2008, 2009, 2010, 2011, 2012]
oscar_revenue = [1005, 170, 427, 133, 232]  # in $M

plt.plot(oscar_years,oscar_revenue)
plt.title("Oscar Movie Revenue")
plt.xlabel("Years")
plt.ylabel("score")

```

```python
Nudmienon_oscar_movies = ["Slumdog Millionaire","Avatar","Inception","Hugo","Lincoln"]
non_oscar_years = [2008, 2009, 2010, 2011, 2012]
non_oscar_revenue = [378, 2788, 829, 185, 275]  # in $M

plt.plot(oscar_years,oscar_revenue, label="oscar movies")
plt.plot(non_oscar_years, non_oscar_revenue,label="non-oscar movies" )
plt.xlabel("Years") 
plt.ylabel("Revenue")
plt.title("Movies analysis")
plt.legend()
```
