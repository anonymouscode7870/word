```python
from sklearn.svm import SVC
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split
```

```python
df = datasets.load_iris(as_frame=True).frame
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
# split data

X= df.drop("target", axis = 1)
y = df["target"]
```

```python
X.head()
```

```python
y.head()
```

```python
# it is very important to scale the data
# in case of SVC we only scale input data not output
# in case of svr we do cale both i/p and o/p

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)
```

```python
X_train.head()
```

```python
# scaling
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
model = SVC()
model.fit(X_train_scaled, y_train)
```

```python
y_pred = model.predict(X_test_scaled)
```

```python
from sklearn.metrics import accuracy_score, classification_report
print(f"Accuracy {accuracy_score(y_test, y_pred)}")
print(f"classification_report {classification_report(y_test, y_pred)}")
```

```python
# lets try with other kernal
model = SVC(kernel="linear") # by default RBF
model.fit(X_train_scaled, y_train)

y_pred = model.predict(X_test_scaled)

print(f"Accuracy {accuracy_score(y_test, y_pred)}")
```

```python
# lets try with other kernal
model = SVC(kernel="poly") # by default RBF
model.fit(X_train_scaled, y_train)

y_pred = model.predict(X_test_scaled)

print(f"Accuracy {accuracy_score(y_test, y_pred)}")
```

```python
# lets try with other kernal
model = SVC(kernel="sigmoid") # by default RBF
model.fit(X_train_scaled, y_train)

y_pred = model.predict(X_test_scaled)

print(f"Accuracy {accuracy_score(y_test, y_pred)}")
```

```python
# lets try with other kernal
c = [0.2, 0.4, 0.7, 1, 2, 3, 4, 6, 7, 8, 9]
for c_val in c:
    model = SVC(C = c_val,kernel="rbf") # by default RBF and c = 1
    model.fit(X_train_scaled, y_train)
    
    y_pred = model.predict(X_test_scaled)
    
    print(f"C = {c_val}, Accuracy {accuracy_score(y_test, y_pred)}")
```
